```python
import json
import glob
import numpy as np
```

```python
input_folder = "../output/output_detect_1k/"
# input_folder = "../output/"

# files = glob.glob(input_folder + '*.json', recursive = False)

successful_urls = set()
for w in open(input_folder+"_successful_sites.txt", "r", encoding="utf-8").read().splitlines():
    # Remove commented and empty lines
    if not w.startswith("#") and not w.strip() == '':
        url = w.lower()
        if url.startswith("https://"):
            url = url[len("https://"):]
        elif url.startswith("http://"):
            url = url[len("http://"):]
        successful_urls.add(url)
successful_urls = list(successful_urls)
```

```python
results_folder = "../notebooks/results/"
version = "0"
```

```python
def get_crawl_time(data):
    start_time = data.get("general", dict()).get("start_time", 0)
    end_time = data.get("general", dict()).get("end_time", 0)

    if start_time == 0 or end_time == 0:
        return None
    
    return end_time - start_time

def get_has_form(data):
    filled_values = data.get("filled_values", [])
    return len(filled_values) != 0
```

```python
crawl_count = 0
form_count = 0
no_form_count = 0

total_crawl_time = 0
total_form_time = 0
total_no_form_time = 0

crawl_times = []
form_times = []
no_form_times = []

for file in successful_urls:
    # Load the data from the .json file
    f = open(input_folder+file+".json", encoding="utf-8") 
    data = json.load(f)
    f.close()

    crawl_time = get_crawl_time(data)
    has_form = get_has_form(data)

    if crawl_time is None:
        continue

    crawl_count += 1
    total_crawl_time += crawl_time
    crawl_times.append(crawl_time)

    if has_form:
        form_count += 1
        total_form_time += crawl_time
        form_times.append(crawl_time)
    else:
        no_form_count += 1
        total_no_form_time += crawl_time
        no_form_times.append(crawl_time)
    
    if not crawl_count % 10:
        print(crawl_count)

print("Calculating results")
avg_crawl_time = total_crawl_time/crawl_count
avg_form_time = total_form_time/form_count
avg_no_form_time = total_no_form_time/no_form_count

crawl_99th_percentile = np.percentile(crawl_times, 99)
form_99th_percentile = np.percentile(form_times, 99)
no_form_99th_percentile = np.percentile(no_form_times, 99)

max_crawl_time = max(crawl_times)
max_form_time = max(form_times)
max_no_form_time = max(no_form_times)

results = {
      "count_crawl": crawl_count,
      "count_form": form_count,
      "count_no_form": no_form_count,
      "avg_crawl_time": avg_crawl_time,
      "avg_form_time": avg_form_time,
      "avg_no_form_time": avg_no_form_time,
      "99th_percentile_crawl": crawl_99th_percentile,
      "99th_percentile_form": form_99th_percentile,
      "99th_percentile_no_form": no_form_99th_percentile,
      "max_crawl_time": max_crawl_time,
      "max_form_time": max_form_time,
      "max_no_form_time": max_no_form_time
}

with open(results_folder + "timing_results_" + version + ".json", 'w', encoding='utf-8') as fp:
        json.dump(results, fp, default=tuple, sort_keys=True, indent=4)
        fp.close()
```
